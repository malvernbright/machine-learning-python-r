```python
X_multi = df.drop('price', axis=1) # X_multi is a new DataFrame that contains all the columns of df except for the 'price' column, which is the target variable we want to predict. The drop() method is used to remove the 'price' column from df, and axis=1 specifies that we want to drop a column (as opposed to a row).
```

```python
X_multi.head()
```

```python
y_multi = df['price'] # y_multi is a Series that contains the 'price' column from df, which is the target variable we want to predict in our multiple linear regression model. By assigning df['price'] to y_multi, we are creating a new variable that we can use as the dependent variable in our regression analysis.
```

```python
y_multi.head()
```

```python
X_multi_cons = sn.add_constant(X_multi) # adds a constant term to the predictor variables, which is necessary for the OLS regression model to include an intercept in the regression equation.
```

```python
X_multi_cons.head()
```

```python
lm_multi = sn.OLS(y_multi, X_multi_cons).fit() # lm_multi is the fitted model object that contains the results of the multiple linear regression analysis. The OLS() function from the statsmodels library is used to fit the model, where y_multi is the dependent variable (the target variable we want to predict) and X_multi_cons is the independent variable (the predictor variables with a constant term added).
```

```python
lm_multi.summary() # The summary() method is called on the fitted model object lm_multi to display a comprehensive summary of the regression results, including statistical measures such as coefficients, standard errors, t-values, p-values, and R-squared values, which help in interpreting the significance and strength of the relationships between the independent variables and the dependent variable in the multiple linear regression model.
```

### Test Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_multi, y_multi, test_size=0.2, random_state=0) # The train_test_split() function from the sklearn library is used to split the dataset into training and testing sets. X_multi and y_multi are the independent and dependent variables, respectively. test_size=0.2 indicates that 20% of the data will be used for testing, while the remaining 80% will be used for training. random_state=0 ensures that the split is reproducible, meaning that the same random split will be generated each time the code is run.
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
lm_a = LinearRegression()
```

```python
lm_a.fit(X_train, y_train) # The fit() method is called on the LinearRegression object lm_a to train the model using the training data (X_train and y_train). This method estimates the coefficients of the linear regression model based on the provided training data, allowing the model to learn the relationships between the independent variables (X_train) and the dependent variable (y_train). After fitting the model, it can be used to make predictions on new data or evaluate its performance on the test set.
```

```python
y_test_a = lm_a.predict(X_test) # The predict() method is called on the fitted LinearRegression model lm_a to make predictions on the test set (X_test). This method uses the learned coefficients from the training phase to generate predicted values for the dependent variable based on the independent variables in X_test. The resulting y_test_a contains the predicted values for the test set, which can be compared to the actual values (y_test) to evaluate the performance of the model.
```

```python
y_train_a = lm_a.predict(X_train) # The predict() method is called on the fitted LinearRegression model lm_a to make predictions on the training set (X_train). This method uses the learned coefficients from the training phase to generate predicted values for the dependent variable based on the independent variables in X_train. The resulting y_train_a contains the predicted values for the training set, which can be compared to the actual values (y_train) to evaluate the performance of the model on the training data.
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_test_a) # The r2_score() function from the sklearn.metrics module is used to calculate the R-squared (coefficient of determination) score for the test set predictions. It compares the actual values (y_test) with the predicted values (y_test_a) to determine how well the model explains the variance in the dependent variable. A higher R-squared score indicates a better fit of the model to the data, with a score of 1 indicating a perfect fit and a score of 0 indicating that the model does not explain any of the variance in the dependent variable.
```

```python
r2_score(y_train, y_train_a) # The r2_score() function is used again to calculate the R-squared score for the training set predictions. It compares the actual values (y_train) with the predicted values (y_train_a) to evaluate how well the model fits the training data. A higher R-squared score indicates a better fit of the model to the training data, with a score of 1 indicating a perfect fit and a score of 0 indicating that the model does not explain any of the variance in the dependent variable for the training set.
```
