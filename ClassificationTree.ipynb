```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Movie_classification.csv', header=0)
```

```python
df.head()
```

```python
df.info()
```

## Missing Value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(df['Time_taken'].mean(), inplace=True)
```

```python
df.info()
```

## Dummy Variable Creation

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['3D_available', 'Genre'], drop_first=True)
```

```python
df.head()
```

## X-y split

```python
X = df.loc[:,df.columns != 'Start_Tech_Oscar']
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
y = df['Start_Tech_Oscar']
type(y)
```

```python
y.head()
```

```python
y.shape
```

## Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

## Training Classification Tree

```python
from sklearn import tree
```

```python
clftree = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clftree.fit(X_train, y_train)
```

## Predict Values using trained model

```python
y_train_pred = clftree.predict(X_train)
y_test_pred = clftree.predict(X_test)
```

```python
y_test_pred
```

## Model Performance Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

## Plotting Decision Tree

```python
dot_data = tree.export_graphviz(clftree, out_file=None, feature_names=X_train.columns, filled=True)
```

```python
from IPython.display import Image
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

## Controlling Tree Growth

```python
clftree2 = tree.DecisionTreeClassifier(max_depth=4, min_samples_leaf=20)
clftree2.fit(X_train, y_train)
dot_data2 = tree.export_graphviz(clftree2, out_file=None, feature_names=X_train.columns, filled=True)
graph2 = pydotplus.graph_from_dot_data(dot_data2)
Image(graph2.create_png())
```

```python
accuracy_score(y_test, clftree2.predict(X_test))
```

## Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf = RandomForestClassifier(n_estimators=250, random_state=42)
```

```python
params_grid = {"max_features": [4,5,6,7,8,9,10],
               "min_samples_split": [2,3,10]}
```

```python
grid_search = GridSearchCV(estimator=rf_clf, param_grid=params_grid, cv=5, n_jobs=-1, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
cvrf_clf = grid_search.best_estimator_
```

```python
accuracy_score(y_test, cvrf_clf.predict(X_test))
```

```python
confusion_matrix(y_test, cvrf_clf.predict(X_test))
```

### Parameter Tuning
1. General Parameters: Guide the overall functioning
2. Booster Parameters: Guide the individual booster (tree/regression) at each step
3. Learning Parameters: Guide the optimization performed

```python
xgb_clf = xgb.XGBClassifier(max_depth=5, n_estimators=1000, learning_rate=.03, n_jobs=-1)
```

```python
xgb_clf.fit(X_train, y_train)
```

```python
accuracy_score(y_test, xgb_clf.predict(X_test))
```

```python
xgb.plot_importance(xgb_clf)
```

```python
xgb_clf = xgb.XGBClassifier(n_estimators=500, learning_rate=.01, random_state=42)
```

```python
param_test1 = {
    'max_depth': range(3,10,2),
    'gamma': [.1, .2, .3],
    'subsample': [.8, .9],
    'reg_alpha': [1e-2, .1, 1]
}
```

```python
grid_search = GridSearchCV(xgb_clf, param_test1,
                           n_jobs=-1, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
cvxg_clf = grid_search.best_estimator_
```

```python
accuracy_score(y_test, cvxg_clf.predict(X_test))
```
