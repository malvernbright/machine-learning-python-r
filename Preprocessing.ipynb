# Simple Linear Regression

```python
import statsmodels.api as sn
```

```python
X = sn.add_constant(df['room_num'])
```

```python
lm = sn.OLS(df['price'], X).fit()
```

```python
lm.summary()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = df['price']
```

```python
X = df[['room_num']]
```

```python
lm2.fit(X, y)
```

```python
print(lm2.intercept_, lm2.coef_)
```
