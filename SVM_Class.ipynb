```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Movie_regression.csv', header=0)
```

```python
df.head()
```

```python
df.info()
```

## Missing Value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(df['Time_taken'].mean(), inplace=True)
```

```python
df.info()
```

## Dummy variables Creation

```python
df = pd.get_dummies(df, columns=['3D_available', 'Genre'], drop_first=True)
```

```python
df.head()
```

## X-y split

```python
X = df.loc[:, df.columns != 'Collection']
y = df['Collection']
type(X), type(y)
```

## Test Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_test_std
```

## Training SVM

```python
from sklearn import svm
```

```python
clf_svm_1 = svm.SVC(kernel='linear', C=.01)
clf_svm_1.fit(X_train_std, y_train)
```

### Predict values using trained model

```python
y_test_pred = clf_svm_1.predict(X_test_std)
y_train_pred = clf_svm_1.predict(X_train_std)
```

```python
y_test_pred
```

### Model Performance

```python
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, confusion_matrix
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
clf_svm_1.n_support_
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'C': [.001, .005, .01, .05, .1, .5, 1, 5, 10, 50, 100, 500, 1000]}
```

```python
clf_svm_1 = svm.SVC(kernel='poly')
```

```python
svm_grid_lin = GridSearchCV(clf_svm_1, params, n_jobs=-1,
                            cv=10, verbose=1, scoring='accuracy')
```

```python
svm_grid_lin.fit(X_train_std, y_train)
```

```python
svm_grid_lin.best_params_
```

```python
linsvm_clf = svm_grid_lin.best_estimator_
```

```python
accuracy_score(y_test, linsvm_clf.predict(X_test_std))
```

### Polynomial kernel

```python
clf_svm_p3 = svm.SVC(kernel='poly', degree=2, C=.01)
clf_svm_p3.fit(X_train_std, y_train)
```

```python
y_train_pred = clf_svm_p3.predict(X_train_std)
y_test_pred = clf_svm_p3.predict(X_test_std)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
clf_svm_p3.n_support_
```

### Radiall

```python
clf_svm_r = svm.SVC(kernel='rbf', C=10)
clf_svm_r.fit(X_train_std, y_train)
```

```python
y_train_pred = clf_svm_r.predict(X_train_std)
y_test_pred = clf_svm_r.predict(X_test_std)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
clf_svm_r.n_support_
```

### Radial Grid

```python
params = {'C': (.001, .005, .01, .05, .1, .5, 1, 5, 10, 50),
          'gamma': (.001, .01, .1, .5, 1)}
```

```python
clf_svm_r = svm.SVC(kernel='rbf')
```

```python
svm_grid_rad = GridSearchCV(clf_svm_r, params, n_jobs=-1,
                          cv=3, verbose=1, scoring='accuracy')
```

```python
svm_grid_rad.best_params_
```

```python
radsvm_clf = svm_grid_rad.best_estimator_
```
