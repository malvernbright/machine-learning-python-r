### Test Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_multi, y_multi, test_size=0.2, random_state=0) # The train_test_split() function from the sklearn library is used to split the dataset into training and testing sets. X_multi and y_multi are the independent and dependent variables, respectively. test_size=0.2 indicates that 20% of the data will be used for testing, while the remaining 80% will be used for training. random_state=0 ensures that the split is reproducible, meaning that the same random split will be generated each time the code is run.
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
lm_a = LinearRegression()
```

```python
lm_a.fit(X_train, y_train) # The fit() method is called on the LinearRegression object lm_a to train the model using the training data (X_train and y_train). This method estimates the coefficients of the linear regression model based on the provided training data, allowing the model to learn the relationships between the independent variables (X_train) and the dependent variable (y_train). After fitting the model, it can be used to make predictions on new data or evaluate its performance on the test set.
```

```python
y_test_a = lm_a.predict(X_test) # The predict() method is called on the fitted LinearRegression model lm_a to make predictions on the test set (X_test). This method uses the learned coefficients from the training phase to generate predicted values for the dependent variable based on the independent variables in X_test. The resulting y_test_a contains the predicted values for the test set, which can be compared to the actual values (y_test) to evaluate the performance of the model.
```

```python
y_train_a = lm_a.predict(X_train) # The predict() method is called on the fitted LinearRegression model lm_a to make predictions on the training set (X_train). This method uses the learned coefficients from the training phase to generate predicted values for the dependent variable based on the independent variables in X_train. The resulting y_train_a contains the predicted values for the training set, which can be compared to the actual values (y_train) to evaluate the performance of the model on the training data.
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_test_a) # The r2_score() function from the sklearn.metrics module is used to calculate the R-squared (coefficient of determination) score for the test set predictions. It compares the actual values (y_test) with the predicted values (y_test_a) to determine how well the model explains the variance in the dependent variable. A higher R-squared score indicates a better fit of the model to the data, with a score of 1 indicating a perfect fit and a score of 0 indicating that the model does not explain any of the variance in the dependent variable.
```

```python
r2_score(y_train, y_train_a) # The r2_score() function is used again to calculate the R-squared score for the training set predictions. It compares the actual values (y_train) with the predicted values (y_train_a) to evaluate how well the model fits the training data. A higher R-squared score indicates a better fit of the model to the training data, with a score of 1 indicating a perfect fit and a score of 0 indicating that the model does not explain any of the variance in the dependent variable for the training set.
```

## Ridge Lasso

```python
from sklearn import preprocessing
```

```python
scalar = preprocessing.StandardScaler().fit(X_train) # The StandardScaler() function from the sklearn.preprocessing module is used to standardize the features in the training set (X_train). The fit() method is called on the StandardScaler object to compute the mean and standard deviation of each feature in X_train, which will be used for scaling the data. This step is important to ensure that all features are on the same scale, which can improve the performance of certain machine learning algorithms that are sensitive to feature scaling. After fitting the scaler, it can be applied to both the training and test sets to transform the features accordingly.
```

```python
X_train_s = scalar.transform(X_train) # The transform() method is called on the fitted StandardScaler object (scalar) to scale the features in the training set (X_train). This method applies the standardization transformation to each feature in X_train using the mean and standard deviation computed during the fit() step. The resulting x_train_scaled contains the standardized features for the training set, which can be used for training machine learning models that require feature scaling.
```

```python
X_test_s = scalar.transform(X_test) # The transform() method is called again on the fitted StandardScaler object (scalar) to scale the features in the test set (X_test). This ensures that the same scaling transformation applied to the training set is also applied to the test set, maintaining consistency in feature scaling between the two datasets. The resulting x_test_scaled contains the standardized features for the test set, which can be used for making predictions and evaluating the performance of machine learning models that require feature scaling. 
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha=0.5) # The Ridge() function from the sklearn.linear_model module is used to create a Ridge regression model object (lm_r) with a specified regularization strength (alpha=0.5). Ridge regression is a linear regression technique that includes a regularization term to prevent overfitting by adding a penalty to the coefficients of the model. The alpha parameter controls the strength of the regularization, with higher values leading to stronger regularization and smaller coefficients.
```

```python
lm_r.fit(X_train_s, y_train) # The fit() method is called on the Ridge regression model object (lm_r) to train the model using the standardized training data (X_train_s and y_train). This method estimates the coefficients of the Ridge regression model based on the provided training data, allowing the model to learn the relationships between the independent variables (X_train_s) and the dependent variable (y_train) while applying regularization to prevent overfitting. After fitting the model, it can be used to make predictions on new data or evaluate its performance on the test set.
```

```python
r2_score(y_test, lm_r.predict(X_test_s)) # The r2_score() function is used to calculate the R-squared score for the Ridge regression model's predictions on the test set. The predict() method of the fitted Ridge regression model (lm_r) is called with the standardized test data (X_test_s) to generate predicted values for the dependent variable. The r2_score() function then compares these predicted values with the actual values (y_test) to evaluate how well the Ridge regression model explains the variance in the dependent variable for the test set. A higher R-squared score indicates a better fit of the model to the data.
```

```python
from sklearn.model_selection import validation_curve
```

```python
validation_curve?
```

```python
param_range = np.logspace(-2, 8, 100) # The np.logspace() function is used to generate a range of values for the regularization parameter (alpha) in Ridge regression. The parameters -2 and 8 specify the exponent range for the logarithmic scale, meaning that the generated values will be between 10^-2 and 10^8. The third parameter, 100, indicates that 100 evenly spaced values will be generated within this range. This range of alpha values can be used to evaluate the performance of the Ridge regression model across different levels of regularization using techniques such as validation curves or grid search.
```

```python
param_range
```

```python
train_scores, test_scores = validation_curve(Ridge(), X_train_s, y_train, param_name='alpha', param_range=param_range, scoring='r2') # The validation_curve() function from the sklearn.model_selection module is used to evaluate the performance of the Ridge regression model across a range of values for the regularization parameter (alpha). The function takes the Ridge() model, standardized training data (X_train_s and y_train), the name of the parameter to vary ('alpha'), the range of parameter values (param_range), and the scoring metric ('r2') as inputs. It returns two arrays: train_scores, which contains the R-squared scores for the training set at each value of alpha, and test_scores, which contains the R-squared scores for the test set at each value of alpha. These scores can be used to analyze how the model's performance changes with different levels of regularization and to select an optimal value for alpha.
```

```python
print(train_scores)
print("---------------test_scores-----------------")
print(test_scores)
```

```python
train_mean = np.mean(train_scores, axis=1) # The np.mean() function is used to calculate the mean of the training scores across the different values of alpha. The axis=1 parameter specifies that the mean should be calculated along the columns (i.e., for each value of alpha), resulting in a 1D array (train_mean) that contains the average R-squared score for the training set at each value of alpha.
test_mean = np.mean(test_scores, axis=1) # The np.mean() function is used to calculate the mean of the test scores across the different values of alpha. The axis=1 parameter specifies that the mean should be calculated along the columns (i.e., for each value of alpha), resulting in a 1D array (test_mean) that contains the average R-squared score for the test set at each value of alpha.
```

```python
train_mean
```

```python
test_mean
```

```python
max(test_mean)
```

```python
sns.jointplot(x=np.log(param_range), y=test_mean) # The sns.jointplot() function from the seaborn library is used to create a joint plot that visualizes the relationship between the logarithm of the regularization parameter values (param_range) and the corresponding mean R-squared scores for the test set (test_mean). The x-axis represents the log-transformed alpha values, while the y-axis represents the mean R-squared scores. This plot can help identify trends in model performance across different levels of regularization and assist in selecting an optimal alpha value for Ridge regression.
```

```python
np.where(test_mean == max(test_mean)) # The np.where() function is used to find the index of the maximum value in the test_mean array. By comparing test_mean with its maximum value (max(test_mean)), we can identify the index or indices where the maximum R-squared score occurs. This information can be useful for determining which value of the regularization parameter (alpha) corresponds to the best performance of the Ridge regression model on the test set.
```

```python
# find the np.where(test_mean == max(test_mean))
param_range[31] # The index 31 is used to access the value of alpha in the param_range array that corresponds to the maximum R-squared score found in the test_mean array. By using this index, we can identify the specific value of the regularization parameter (alpha) that yields the best performance for the Ridge regression model on the test set.
```

```python
lm_r_best = Ridge(alpha=param_range[31]) # The Ridge regression model is instantiated with the optimal value of alpha (param_range[31]) that corresponds to the maximum R-squared score found in the test_mean array. This model (lm_r_best) is now configured to use the best regularization strength for making predictions and evaluating performance on the test set.
```

```python
lm_r_best.fit(X_train_s, y_train) # The fit() method is called on the Ridge regression model object (lm_r_best) to train the model using the standardized training data (X_train_s and y_train). This method estimates the coefficients of the Ridge regression model based on the provided training data, allowing the model to learn the relationships between the independent variables (X_train_s) and the dependent variable (y_train) while applying regularization with the optimal alpha value. After fitting the model, it can be used to make predictions on new data or evaluate its performance on the test set.
```

```python
r2_score(y_test, lm_r_best.predict(X_test_s)) # The r2_score() function is used to calculate the R-squared score for the Ridge regression model (lm_r_best) with the optimal alpha value. The predict() method of the fitted model is called with the standardized test data (X_test_s) to generate predicted values for the dependent variable. The r2_score() function then compares these predicted values with the actual values (y_test) to evaluate how well the Ridge regression model with the optimal regularization strength explains the variance in the dependent variable for the test set. A higher R-squared score indicates a better fit of the model to the data.
```

```python
r2_score(y_train, lm_r_best.predict(X_train_s)) # The r2_score() function is used to calculate the R-squared score for the Ridge regression model (lm_r_best) with the optimal alpha value. The predict() method of the fitted model is called with the standardized training data (X_train_s) to generate predicted values for the dependent variable. The r2_score() function then compares these predicted values with the actual values (y_test) to evaluate how well the Ridge regression model with the optimal regularization strength explains the variance in the dependent variable for the training set. A higher R-squared score indicates a better fit of the model to the data.
```
